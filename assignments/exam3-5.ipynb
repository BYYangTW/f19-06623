In the numerical integration of a differential equation, a predictor-correction scheme is often used to improve the results. In this approach, you start with the current value, and predict the next value. Then, a correction is applied to the prediction to make it more accurate.

Suppose we have an initial value ODE as $y' = f(t, y), y(t_0) = y_0$ and we denote the step size to take as $h$.

We can make a prediction step to obtain an intermediate estimate using the Euler method as:

$\widetilde{y}_{i+1} = y_i + h f(t_i, y_i)$

Then, we make a correction step by using the trapezoid rule as:

$y_{i + 1} = y_i + \frac{1}{2} * h * [f(t_i, y_i) + f(t_{i+1}, \widetilde{y}_{i + 1})]$

Let $y'(t, y) = -y, y(0) = 1.0$.

Implement the predictor-corrector scheme above to find $y(t)$ from $t=0$ to $t=1$ using a step size of $h=0.1$. Compare your answer to the solution found using `scipy.integrate.solve_ivp`.
