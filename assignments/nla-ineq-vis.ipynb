```python
def objective(X):
    x1, x2 = X
    return np.exp(x1) * (4 * x1**2 + 2 * x2**2 + 4 * x1 * x2 + 2 * x2 + 1)
```

$x_1 x_2 - x_1 - x_2 \le -1.5$

and

$x_1 x_2 \ge -10$.
