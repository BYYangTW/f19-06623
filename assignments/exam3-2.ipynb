The following reactions take place in constant volume, gas-phase batch reactor:

$A + B \leftrightharpoons C + D$

$B + C \leftrightharpoons X + Y$

$A + X \leftrightharpoons Z$

The following algebraic equations have been derived to describe the equilibrium of these three reactions. In these equations, $C_i$ represents the equilibrium concentration of species $i$.

$K_{C1} = \frac{C_C C_D}{C_A C_B}$

$K_{C2} = \frac{C_X C_Y}{C_B C_C}$

$K_{C3} = \frac{C_Z} {C_A C_X}$

The following equations were derived from stoichiometric constraints.

$C_A = C_{A0} - C_D - C_Z$

$C_B = C_{B0} - C_D - C_Z$

$C_C = C_D - C_Y$

$C_Y = C_X + C_Z$

Given $C_{A0} = C_{B0} = 1.5$ and $K_{C1} = 1.06$, $K_{C2} = 2.63$ and $K_{C3} = 5$, find the equilibrium concentrations of $C_A, C_B, C_C, C_D, C_X, C_Y, C_Z$.
