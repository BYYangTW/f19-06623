### Minimal cost diet



Foods A and B have 600 and 500 calories, contain 15g and 30g of protein, and cost \\$1.80 and \\$2.10 per unit each. Find the minimum cost diet containing at least 3900 calories and at least 150g of protein.



Show one example around your solution that illustrate that the constraints are not met.



Show another example where the constraints are met, but the objective is not less less than the solution you found.



### Solution



This is an optimization problem with inequality constraints. You just set up the objective to minimize, which is the cost, and one function for each constraint. You could also add positivity constraints, but it is not necessary here.



```python
def objective(X):
    a, b = X
    return a * 1.8 + b * 2.10

def constraint1(X):
    a, b = X
    calories = 600 * a + 500 * b
    return calories - 3900

def constraint2(X):
    a, b = X
    protein = 15 * a + 30 * b
    return protein - 150

from scipy.optimize import minimize
sol = minimize(objective, (1, 1), constraints=[{'type': 'ineq', 'fun': constraint1},
                                               {'type': 'ineq', 'fun': constraint2}])
sol
```

```python
print(f'The minimum cost is ${sol.fun:1.2f} for {sol.x[0]:1.1f} units of A and {sol.x[1]:1.1f} units of B.')
```

It is a good idea to check our constraints. For the first constraint this shows you get exactly 3900 calories.



```python
constraint1(sol.x)
```

And we see that we get 150 g of protein at this solution.



```python
constraint2(sol.x)
```
