For the first part, you can simply solve for V and evaluate it: $V = R T / P$.



```python
B = -388
C = -26000
R = 83.14
T = 200 + 273.15
P = 10

print(f'V = {R * T / P:1.1f} cm^3/mol')
```

Some common mistakes here were not using the best conversion of Celsius to Kelvin (K = C + 273.15). V in this equation is a *molar volume* and should have units of cm^3/mol.

The point of the next



```python
import numpy as np

def f(V):
    return P * V / R / T - 1 - B / V - C / V**2

# Show a single number works
f(1000), f(3000)
```

Here I show that the function works with an array. By works, I only mean it gives numbers as an array, and that no errors are raised.



```python
f(np.array([1000, 3000]))
```

Are errors possible? Certainly, if you pass a list in, then you get an error because we do not make sure it is an array when we run the function.



```python
f([1000, 3000])
```

The point of making the plot is to graphically see where it is equal to zero. This is a coarse grained version that shows there is a zero near 3500 cm^3/mol. We cannot estimate the solution more accurately than that in this graph.



```python
%matplotlib inline
import matplotlib.pyplot as plt

V = np.linspace(3000, 4000)
plt.plot(V, f(V))
plt.xlabel('V (cm^3/mol)')
plt.ylabel('f(V)');
```

To get a more accurate estimate, we make a new, zoomed in plot by changing the x-axis values.



```python
V = np.linspace(3480, 3500)
plt.plot(V, f(V))
plt.xlabel('V (cm^3/mol)')
plt.ylabel('f(V)');
```

The zero is around 3486 cm^3/mol based on the previous graph. This just shows that the function is close to zero at that value.



```python
f(3486)
```

You can get a cubic polynomial by multiplying f(V) by V^2. This is sufficient for answering this question.
