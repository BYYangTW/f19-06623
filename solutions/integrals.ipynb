### even or odd functions



An odd function $o(x)$ satisfies $o(x)=-o(-x)$ for all $x$. Therefore, for any finite $t$:

$$\int_{-t}^{t}o(x) dx=0$$.

An even function $e(x)$ satisfies $e(x) = e(-x)$ for all $x$. Therefore, for any $t$:

$$\int_{-t}^{t}e(x) dx=2 \int_{0}^{t}e(x) dx$$.

### Convenient change of variables



The following statement is true, and is sometimes used to make integrals easier to evaluate.

$\int_a^b f(x)dx = \int_a^b f(a + b - x) dx$

Show that it is true for the integral:

$\int_3^7 \frac{\ln(x+2)}{\ln(24 + 10 x - x^2)}dx$



### Evaluate this integral numerically using Python



$\int_0^\infty \frac{\ln(x)}{x^2 + 2 x + 4}$



### solution



To determine if this function is even or odd, we need to show that for many choices of x, if $f(x) = -f(-x)$  or if $f(x) = f(-x)$, and we need to show the integral. It is not possible to show this for all x, so we have to choose a range of values. Here, I choose the range -1 to 1 for x.



```python
import numpy as np
from scipy.integrate import quad

def f1(x):
    return (x**3 - 2*x) / (x**4 + 1)

x = np.linspace(-1, 1, 5)
print(f1(x), -f1(-x))
```

You can see by inspection that $f(x) = -f(-x)$. It is easier to see this is the case like this.



```python
f1(x) == -f1(-x)
```

And this shows also that all of the entries are True.



```python
np.all(f1(x) == -f1(-x))
```

\#+END\_SRC

The results above suggest that the integral should evaluate to zero, and we show that here.



```python
# Now show the integral value
t = 1
I, err = quad(f1, -t, t)
print(I)
```

That shows that f(x) = -f(-x) and the integral is equal to zero, and $f(x) = -f(-x)$ so the function is odd.

For the next part, first we evaluate the integral as given.



```python
def f2(x):
    return np.log(x + 2) / np.log(24 + 10 * x - x**2)

a = 3
b = 7

quad(f2, a, b)
```

Then we evaluate the integral with the changed variables.



```python
# let y = a + b - x

def f3(y):
    x = a + b - y
    return f2(x)

quad(f3, a, b)
```

You can see by inspection here that they are the same.

Finally, the only thing to do differently here is to use `np.inf` in the integration limits.



```python
def f4(x):
    return np.log(x) / (x**2 + 2 * x + 4)

I, err = quad(f4, 0, np.inf)
print(f'The integral evaluates to {I}.')
```
