### Error estimation in the trapezoid method



The trapezoid method can be used to approximate the integral of a function.

$I = \int_a^b f(x) dx$

Suppose we are given:

$f(x) = e^x \cos x$ and we seek the integral of $f(x)$ from $x=0$ to $x=\pi$ using the trapezoid method. The error in this approximation can be estimated as:

$Err = -\frac{h^2}{12}(f'(b) - f'(a))$

where $h$ is the width of each trapezoid. It is given here that $f'(x) = e^x(\cos(x)-\sin(x))$.

First, use the error estimate above to determine what $h$ is required to get an error less than or equal to 0.001, and then estimate how many points are required in the interval of x=0 to &pi; to achieve that.



Next, use that information with trapezoid method to estimate the integral with at least the desired accuracy.



Finally, use the quad function to determine the integral value and determine if your estimate from the trapezoid method is within the desired error.



### solution



In the first step, we can solve for the h required algebraically from the equation given and the desired error level. Then we simply estimate how many points we need by dividing the range of the interval by that h. You need to round that to the nearest integer for the number of points. It is not essential which direction you round, we are only estimating the number here and being off by one is probably not significant.



```python
import numpy as np

def f(x):
    return np.exp(x) * np.cos(x)

def fp(x):
    return np.exp(x) * (np.cos(x) - np.sin(x))

a = 0
b = np.pi

h = np.sqrt(-0.001 / (fp(b) - fp(a)) * 12)
print(f'h = {h}')

print(f'Estimated number of points is {(b - a) / h}')
```

Based on the answer above, you need about 141 points. To estimate the integral, we just create a linspace array on the integration range with the desired number of points and then evaluate the function on that array and use np.trapz to get the integral.



```python
x = np.linspace(a, b, 141)
y = f(x)
I = np.trapz(y, x)
print(I)
```

To compute the error, we use quad and assume that because the error in quad is so small we can neglect it.



```python
from scipy.integrate import quad
I0, err = quad(f, a, b)
I0
```

Then we just look at the difference.



```python
I0 - I
```

Remarkably, this difference is very close to the error we specified!
